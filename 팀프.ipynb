```python
import numpy as np
import cv2
import os 
import csv
import colorsys
from sklearn.cluster import KMeans

# 데이터 폴더에서 이미지 파일이름 가져오기
file_list=os.listdir("C:/Users/Hong Sumin/Desktop/3-1/data/CHANEL5")
final_list=[]
column=["name","r","g","b","h","s","v"]

# 파일
for file in file_list:
    file_name=file.split(".")
    # 파일이름 가져오기
    file_name=file_name[0]
    file_name=file_name.split(" ")[0]
    print(file_name)
    img=cv2.imread("C:/Users/Hong Sumin/Desktop/3-1/data/CHANEL5/"+file)
    # rgb추출
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    # reshape
    img=img.reshape(img.shape[0]*img.shape[1],3)

    k=10
    # KMeans Clustering
    clt=KMeans(n_clusters=k)
    clt.fit(img)
    
    # center값
    for center in clt.cluster_centers_:
        print(center)
        hsv=colorsys.rgb_to_hsv(center[0],center[1],center[2])
        temp=[]
        # 파일이름 저장
        temp.append(file_name)
        # 센터값 저장
        for i in range(len(center)):
            temp.append(center[i])
        for i in range(len(hsv)):
            temp.append(hsv[i])
        # 최종 리스트에 저장    
        final_list.append(temp)
# csv 파일 저장 - 한글 깨지면 인코딩 바꿔야 할 수도…
file=open("CHANEL5.csv","w",newline='',)
csvfile=csv.writer(file)
for row in final_list:
    csvfile.writerow(row)
file.close()
```

```python
import cv2 
import numpy as np

def tracking(img_name): # img_name = 추출할 이미지 경로 혹은 이름
    img = cv2.imread(img_name) 
    lower_ba = np.array([204, 153, 153]) 
    upper_ba = np.array([255, 204, 204])
  #  ba_range=cv2.inRange(hsv,lower_ba,upper_ba)
    #for i in range(lower_red,upper_red):
    
    #153 102 g
    
   # while True:
        #ret, fram = img.read()
    hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    
    
    red_range = cv2.inRange(hsv,lower_ba,upper_ba)
        
    red_result = cv2.bitwise_and(img,img,mask = red_range)
    red_result = cv2.resize(red_result,(300,300))
       
    img = cv2.resize(img,(300,300))
    
    # 원래 이미지 화면 띄우기
    cv2.imshow("original",img)
    # 추출한 이미지 화면 띄우기
    cv2.imshow("red",red_result)
    
    #추출한 이미지 저장
    cv2.imwrite('result1.PNG', red_result)
        
    cv2.waitKey(0)
        
    cv2.destroyAllWindows()
img_name = "C:/Users/Hong Sumin/Desktop/3-1/hi3.jpg"
tracking(img_name)

print(type(tracking(img_name)))
        
```
